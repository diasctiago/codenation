### Resolução

```python
estado = list(df['estado_residencia'].unique())
```

```python
estatisticas = {}
```

```python
for x in estado:
    media = float(df[df['estado_residencia'] == x].groupby('estado_residencia')['pontuacao_credito'].mean())
    mediana = float(df[df['estado_residencia'] == x].groupby('estado_residencia')['pontuacao_credito'].median())
    moda = float(df[df['estado_residencia'] == x].groupby('estado_residencia')['pontuacao_credito'].agg(pd.Series.mode))
    desvio = float(df[df['estado_residencia'] == x].groupby('estado_residencia')['pontuacao_credito'].std())
    resultado = {'moda': moda, 'mediana': mediana, 'media': media, 'desvio_padrao': desvio}
    estatisticas.update({x: resultado})
```

```python
estatisticas
```

```python
with open('submission.json', 'w') as f:
    json.dump(estatisticas, f)
```
